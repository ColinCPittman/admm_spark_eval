```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D
from matplotlib.patches import Patch


# KSU official colors
ksu_black = "#2D2926"
ksu_gold  = "#FFC629"

# Unique partition values
parts = sorted(df['partitions'].unique())
n = len(parts)

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, n, figsize=(4*n, 5), sharey=False)

# Legend elements
legend_elements = [
    Patch(facecolor=ksu_black, label='Runtime (s)'),
    Line2D([0], [0], marker='o', color=ksu_gold, label='Accuracy (%)',
           markerfacecolor=ksu_gold, markersize=8, linestyle='None')
]

for ax, p in zip(axes, parts):
    sub = df[df['partitions'] == p].reset_index(drop=True)
    labels = sub['algorithm'] + ' v' + sub['version']

    # Black bars for runtime
    ax.bar(labels, sub['runtime_s'], color=ksu_black)
    ax.set_title(f'Partitions = {p}', fontsize=12)
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=45, ha='right')
    if ax is axes[0]:
        ax.set_ylabel('Runtime (s)', fontsize=12)

    # Twin axis for accuracy dots
    ax2 = ax.twinx()
    ax2.scatter(labels, sub['accuracy_pct'], color=ksu_gold, marker='o', s=80, zorder=5)
    y_min = df['accuracy_pct'].min() * 0.98
    y_max = df['accuracy_pct'].max() * 1.01
    ax2.set_ylim(y_min, y_max)
    if ax is axes[0]:
        ax2.set_ylabel('Accuracy (%)', fontsize=12)

# Add legend
fig.legend(handles=legend_elements, loc='upper center', ncol=2,
           frameon=False, bbox_to_anchor=(0.5, 1.12))

fig.suptitle("Runtime and Accuracy by Partition", fontsize=14, y=1.18)
plt.tight_layout()
plt.show()
```
